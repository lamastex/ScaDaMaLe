```
def compute_dst_sq(): # (N, M)
  # dst[n, m] = |x_n - x'_m|^2
  dst = train_data.cartesian(test_data)
  dst = np.zeros((N, M))
  for i in range(N):
    for j in range(M):
      tmp = train_data[i, :] - test_data[j, :]
      dst[i, j] = np.sum(tmp * tmp)
  return dst

def compute_pu(data): # (data.N, T)
  # pu[n, t] = <data_n, ray_t>
  return np.dot(data, rays.T)

dst = compute_dst_sq()
pu_train = compute_pu(train_data)
pu_test = compute_pu(test_data)
```

```
def compute_ray_lengths(): # (M, T)
  lengths = np.zeros((M, T, N))
  # lengths[m, t, n] = dst[n, m] / (2 * (pu_train[n, t] - pu_test[m, t]))
  for n in range(N):
    for m in range(M):
      lengths[m, :, n] = dst[n, m] / (2 * (pu_train[n, :] - pu_test[m, :]))
      
  lengths[lengths < 0] = np.inf
  
  if one_vs_all:
    for n in range(N):
      lengths[n, :, n] = np.inf

  lengths = np.min(lengths, axis=2)
  # lengths = np.sort(lengths, axis=1) # not currently needed
  return lengths

lengths = compute_ray_lengths()
```

```
def _old_compute_integrals(): # (M)
  integrals = np.zeros((M,))
  for m in range(M):
    current = 0.
    cur_coeff = 0.
    cur_T = np.sum(lengths[m, :] < np.inf)
    prev_len = 0.
    
    for t in range(cur_T):
      cur_len = lengths[m, t]
      current += (cur_len - prev_len) * (cur_T - t)
      if t > 0:  # can omit with large D
        current += cur_coeff * (-D + 1) * (np.power(cur_len, -D + 1) - np.power(prev_len, -D + 1))
      cur_coeff += np.power(cur_len, D)
      prev_len = cur_len
    
    integrals[m] = current
  return integrals

def compute_integrals(): # (M)
  integrals = np.zeros((M,))
  finite = np.sum(np.isfinite(lengths), axis=1)
  lengths[np.isinf(lengths)] = 0
  integrals = np.sum(lengths, axis=1) / finite
  return integrals

integrals = compute_integrals()
```
