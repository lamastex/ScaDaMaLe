This is a short description of the idea behind KLL quantile
approximation algorithm described in
https://arxiv.org/pdf/1603.05346.pdf.  
Let X = {x*1, x*2, ..., x*n} be a set of real numbers. The rank R(x) of
any point x in X is defined as the number of x*i ≤ x. The normalized
rank NR(x) = R(x) / n. α-quantile of X is defined as any number x such
that NR(x) = α.

Given X as a stream, we wish to compute a data structure that would
allow us to *estimate* α-quantile of X for any α in the following sense:
we want be able to return an estimate **x'** such that **|x-x'| ≤ en**
with probability **1-d** for some error **e** and confidence **d**.

We do not assume any distribution on X, thus if n is much larger than
the space available to us, we cannot escape the error margin growing
together with n.

The basic idea of the algorithm is as follows: we divide X into n/k
chunks of k elements. Each chunk is processes by a *compactor* that
collects elements and associates an equal weight w with each element.
After k elements are collected, the compactor sorts them and then either
removes the odd-indexed ones or the even-indexed ones (deciding
randomly), sets the weight for all of them to 2w and passes the result
to another compactor. This forms a hierarchy of chained compactors.

As a first step, letting all compactors have the same capacity k yields
an algorithm with error bounded by **(n/k)log(n/k)** and space
complexity bounded by **klog(n/k)**. Setting **k = O( (1/e) log(en) )**
yields an error of **en** with space **O( (1/e) log^2(en) )**. More
precise analysis shows that the space is really **O( (1/e) log(en) )**.

KLL present an improved version of this algorithm where the capacity of
a compactor is determined by its position in the chain. They let the
size exponentially decay towards the lower levels, and show that the
error does not change asymoptotically, while the space usage is clearly
improved.
